```
f = open(data_directory+"ucfTrainTestlist/testlist01.txt", "r")
temp = f.read()
videos = temp.split('\n')

# creating a dataframe having video names
test = pd.DataFrame()
test['video_name'] = videos
```

```
from shutil import copy
for i in tqdm(range(test.shape[0]-1)):
  videoFile = test['video_name'][i]
  path = data_directory+'Videos/'+videoFile.split(' ')[0].split('/')[1]
  cls = videoFile.split(' ')[0].split('/')[1].split('_')[1]
  if not os.path.exists('Videos/test/'+cls):
      os.makedirs('Videos/test/'+cls)
  copy(path,'Videos/test/'+cls)
```
